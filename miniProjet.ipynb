```python
# 1. Régression Linéaire

import numpy as np
from sklearn.linear_model import LinearRegression

# dataset : kilométrage et prix des voitures
x_linear = [5000, 15000, 30000, 45000, 60000] # kilométrage en km
y_linear = [220000, 200000, 180000, 160000, 140000] # prix de voiture en DH

# Scratch 
print("\n--> Régression Linéaire : From Scratch")

# calcule le nombre de points
n = len(x_linear)

# calcule les moyennes de x et y
mean_x = sum(x_linear)/n
mean_y = sum(y_linear)/n

# calcul du numérateur et dénominateur de la pente
num = sum((x_linear[i]-mean_x)*(y_linear[i]-mean_y) for i in range(n))
den = sum((x_linear[i]-mean_x)**2 for i in range(n))

# calcul des paramètres a (pente) et b (intercept)
a = num/den
b = mean_y - a*mean_x

# affichage des résultats
print("Pente =", a)
print("Intercept =", b)


print("Prediction 25000 km =", a*25000 + b) # Je teste le modèle avec une nouvelle valeur


# sklearn 
print("\n--> Régression Linéaire : Sklearn ")

# transformation des données en format numpy 
X = np.array(x_linear).reshape(-1,1)
Y = np.array(y_linear)


model = LinearRegression() # Création du modèle


model.fit(X, Y) # Entraînement du modèle

# affichage des résultats
print("Pente =", model.coef_[0])
print("Intercept =", model.intercept_)

print("Prediction 25000 km =", model.predict([[25000]])[0]) # Prédiction avec sklearn
```

```python
# 2. Régression Multiple

from sklearn.linear_model import LinearRegression

# dataset 
X_multi = [
    [5000, 2020, 2.0],
    [30000, 2018, 1.6],
    [15000, 2019, 2.2],
    [40000, 2017, 1.8]
]# km, année, moteur
y_multi = [220000, 150000, 200000, 130000]# prix de la voiture en DH

# normalisation: réduis les valeurs pour faciliter le calcul
X_multi_scaled = [[x/1000 for x in row[:2]] + [row[2]] for row in X_multi]

# Scratch
print("\n--> Régression Multiple : Scratch ")

# initialisation des poids et du biais
w = [0,0,0]
b = 0

# taux d’apprentissage
lr = 1e-3

# nombre d’itérations
epochs = 50000

# nombre de données
n = len(X_multi_scaled)

# descente de gradient
for _ in range(epochs):
    dw = [0,0,0]
    db = 0
    # parcours toutes les données
    for i in range(n):
        # calcul de la prédiction
        y_pred = sum(w[j]*X_multi_scaled[i][j] for j in range(3)) + b
        # calcul de l'erreur
        error = y_pred - y_multi[i]
        # mise à jour des gradients
        for j in range(3):
            dw[j] += error*X_multi_scaled[i][j]
        db += error
    # mise à jour des poids
    for j in range(3):
        w[j] -= lr*dw[j]/n
    # mise à jour du biais
    b -= lr*db/n

# résultat final
print("Poids =", w)
print("Intercept =", b)


# sklearn 
print("\n--> Régression Multiple : Sklearn ")

# création du modèle
model = LinearRegression()

# entraînement automatique
model.fit(X_multi_scaled, y_multi)

# résultats
print("Poids =", model.coef_)
print("Intercept =", model.intercept_)
```

```python
# 3. Régression Polynomiale

import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# dataset : 
x_poly = [5000, 15000, 30000, 45000, 60000]# kilométrage en km
y_poly = [220000, 200000, 180000, 160000, 140000]# prix de la voiture en DH

# normalisation
x_scaled = [xi/10000 for xi in x_poly]

# création des variables (x, x²)
X_poly = [[xi, xi**2] for xi in x_scaled]

# scratch 
print("\n--> Polynomiale : From Scratch ")

w1, w2, b = 0, 0, 0 # initialisation des paramètres

lr = 1e-3 # taux d’apprentissage

epochs = 100000 # nombre d’itérations

n = len(X_poly) # nombre de données

# descente de gradient
for _ in range(epochs):
    dw1 = 0
    dw2 = 0
    db = 0
    for i in range(n):
        x1, x2 = X_poly[i]
        y_pred = w1*x1 + w2*x2 + b # calcul de la prédiction
        error = y_pred - y_poly[i] # calcul de l'erreur
        # calcul des gradients
        dw1 += error*x1
        dw2 += error*x2
        db += error
    # mise à jour des paramètres
    w1 -= lr*dw1/n
    w2 -= lr*dw2/n
    b -= lr*db/n

# résultat final
print("w1 =", w1)
print("w2 =", w2)
print("b =", b)


# sklearn 
print("\n--> Polynomiale : sklearn")

# transformation des données
x_arr = np.array(x_scaled).reshape(-1,1)

# création des variables polynomiales
poly = PolynomialFeatures(degree=2)
X_poly_sklearn = poly.fit_transform(x_arr)

# création du modèle
model = LinearRegression()

# entrainement
model.fit(X_poly_sklearn, y_poly)

# Résultats
print("Coefficients =", model.coef_)
print("Intercept =", model.intercept_)
```
